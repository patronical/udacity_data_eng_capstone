### Scraped Twitter Following


```python
# import libraries
import got3
import pandas as pd
import pickle
import itertools
import os
os.getcwd()
```

#### Import following files

```python
# These are the accounts the seeds are following
# These files are generated by mininig twitter seed accounts
with open('following.data', 'rb') as filehandle:  
    # read the data as binary data stream
    fdata = pickle.load(filehandle)
# build a list of lists
fd_list = [fdata]
for i in range(51):
    fh = 'following{}.data'.format(i)
    with open(fh, 'rb') as filehandle:  
        # read the data as binary data stream
        fdata = pickle.load(filehandle)
        fd_list.append(fdata)
    
len(fd_list)
```

```python
# flatten the list of lists

followflat = list(itertools.chain.from_iterable(fd_list))
len(followflat)
```

```python
# count up the followings
from collections import Counter
fCount = Counter(followflat)

print(fCount.most_common(10))
```

```python
# we want to mine a bunch of accounts that are commonly followed
highC = [k[0] for k in fCount.most_common(250)]
```

#### Import seeds already mined

```python
# this is the list of the accounts manually scraped
# these are the first itteration of mined accounts
df_seed = pd.read_csv('crypto_seed.csv')

seeds = df_seed['seed'].tolist()
print(seeds)
```

```python
# remove end space on names on seeds
for i in range(len(seeds)):
    if seeds[i][-1] == ' ':
        seeds[i] = seeds[i][:-1]
print(seeds)
```

```python
# These are the second itteration of mined accounts 
with open('newseed0.data', 'rb') as filehandle:  
    # read the data as binary data stream
    newseeds = pickle.load(filehandle)
print(newseeds)
```

```python
# these are the first and second itteration of mined accounts
mined_seeds = seeds + newseeds
print(mined_seeds)
```

#### Find New Seeds to Mine

```python
# this is the list for third itteration mining
NewSeed = [k for k in highC if k not in mined_seeds]
len(NewSeed)
```

```python
print(NewSeed)
```

```python
# pickle the list for mining
with open('newseed1.data', 'wb') as filehandle:  
    # store the data as binary data stream
    pickle.dump(NewSeed, filehandle)
```

```python
# 1st, 2nd, 3rd itteration twitter accounts mined for tweets
all_seeds = (mined_seeds + NewSeed)
all_seeds.sort(key=str.lower)
print(all_seeds)
```

```python
# save for maintenance mining later
with open('crypto_tweeter_seeds_060419.pkl', 'wb') as filehandle:  
    # store the data as binary data stream
    pickle.dump(all_seeds, filehandle)
```
